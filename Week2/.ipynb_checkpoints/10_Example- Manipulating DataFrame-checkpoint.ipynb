```python
import pandas as pd
```

```python
df = pd.read_csv('presidents.csv')
df.head()
```

```python
df['First'] = df['President']
df['First'] = df['First'].replace('[ ].*',"", regex=True)
df.head()
```

```python
# but that was inefficent lets delete it and use apply() instead
del df['First']
df.head()
```

The apply() function on a dataframe will take some arbitrary function you have written and apply it to either a Series (a single column) or DataFrame across all rows or columns. Lets write a function which just splits a string into two pieces using a single row of data -

```python
def splitname(dataframe):
    dataframe['First'] = dataframe['President'].split(' ')[0]
    dataframe['Last'] = dataframe['President'].split(' ')[-1]
    return dataframe

df = df.apply(splitname, axis=1)
df.head()
```

```python
del df['First']
del df['Last']
df.head()
```

extract() - 

```python
pattern = '(?P<First>^[\w]*).* (?P<Last>[\w]*$)'
names = df['President'].str.extract(pattern)
names.head()
```

```python
df['First'] = names['First']
df['Last'] = names['Last']
df.head()
```

```python
df['Born'] = df['Born'].str.extract('([\w]{3} [\w]{1,2}, [\w]{4})')
df.head()
```

```python
df['Born'] = pd.to_datetime(df['Born']) # changing data type to datetime
df['Born'].head()
```
