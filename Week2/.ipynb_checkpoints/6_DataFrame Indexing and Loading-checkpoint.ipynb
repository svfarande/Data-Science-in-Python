```python
# Now if you want to set Serial no. as index then we can use index_col
df = pd.read_csv('Admission_Predict.csv', index_col=0)
df.head()
```

If you want to change column names in Pandas, we can use the rename() function. It takes a parameter called columns, and we need to pass into a dictionary with the keys as the old column name and the value as the corresponding new column name -

```python
new_df=df.rename(columns=
                 {'GRE Score':'GRE Score', 'TOEFL Score':'TOEFL Score',
                  'University Rating':'University Rating', 
                  'SOP': 'Statement of Purpose',
                  'LOR': 'Letter of Recommendation',
                  'CGPA':'CGPA', 'Research':'Research',
                  'Chance of Admit':'Chance of Admit'})
new_df.head() 
```

Look LOR didn't changed because LOR doesn't exists and its 'LOR '. Same goes with chance of admit the last column -

```python
df.columns
```

```python
new_df = df.rename(columns = {'LOR ':'Letter of Recommendation'})
new_df
```

We can do same using python string function strip(), lets see what is does to our last column chance of admit -

```python
new_df.columns
```

```python
new_df = new_df.rename(mapper=str.strip, axis=1)
new_df.columns
```

```python
# Every thing we did till now was on copy of df and stored as new_df
df.columns
```

This space is common while loading such data, instead we can use our own column name, let say we want all UPPER case -

```python
cols = df.columns
cols = [x.upper().strip() for x in cols]
df.columns = cols
df.columns
```
