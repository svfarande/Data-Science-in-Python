# Pivot Table

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cwurData.csv')
df.head()
```

```python
def create_category(rank):
    if (rank>=1) & (rank<=100):
        return "First Tier Top University"
    elif (rank>=101) & (rank<=200):
        return "Second Tier Top University"
    elif (rank>=201) & (rank<=300):
        return "Third Tier Top University"
    return "Other Top University"

df['Rank_Level'] = df['world_rank'].apply(lambda x:create_category(x))
df.head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', 
               aggfunc=[np.mean]).head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', 
               aggfunc=[np.mean, np.max]).head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', 
               aggfunc=[np.mean, np.max], margins=True).head()
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', 
               aggfunc=[np.mean, np.max], margins=True).tail()
```

```python
new_df = df.pivot_table(values='score', index='country', columns='Rank_Level', 
                        aggfunc=[np.mean, np.max], margins=True)
print(new_df.index)
print(new_df.columns)
```

+ UnStacking - is pivoting innermost row index to lowermost column index
