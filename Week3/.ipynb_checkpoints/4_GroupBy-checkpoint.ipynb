### GroupBy 
Its is like Split - Apply - Combine in Pandas

```python
import pandas as pd
import numpy as np
```

### Splitting 

```python
df = pd.read_csv('census.csv')
df = df[df['SUMLEV']==50]
df.head()
```

```python
%%timeit -n 3

for s in df['STNAME'].unique():
    avg = np.average(df[df['STNAME']==s]['CENSUS2010POP'])
    # print(s + ' has average population of ' + str(avg))
    # Output -
    # Alabama has average population of 71339.34328358209
    # Alaska has average population of 24490.724137931036...
```

```python
%%timeit -n 3

for group, frame in df.groupby('STNAME'):
    avg = np.average(frame['CENSUS2010POP'])
    # print(group + ' has average population of ' + str(avg))
    # Output -
    # Alabama has average population of 71339.34328358209
    # Alaska has average population of 24490.724137931036...
```

If no column identifier is given to groupby() it acts on index -

```python
df = df.set_index('STNAME')

def set_batch_number(item):
    if item[0] < 'M':
        return 0
    if item[0] < 'Q':
        return 1
    return 2

for group,frame in df.groupby(set_batch_number):
    print('There are '+ str(len(frame)) + ' records in group ' 
          + str(group) + ' for processing')
```

```python
df = pd.read_csv('listings.csv')
df.head()
```

```python
df = df.set_index(['cancellation_policy','review_scores_value'])

for group,frame in df.groupby(level=(0,1)):
    print(group)
```

```python
def grouping_fun(item):
    if item[1] == 10.0:
        return (item[0], '10.0')
    else:
        return (item[0], 'not 10.0')

for group, frame in df.groupby(by=grouping_fun):
    print(group)
```

```python
df.head()
```

### Aggregation

```python
df = df.reset_index()
df.groupby('cancellation_policy').agg({'review_scores_value':np.average})
```

__note__ - in above agg function np.average() doesn't work because it doesn't excludes NaN. Hence we need to use np.nanmean() which excludes NaN

```python
df.groupby('cancellation_policy').agg({'review_scores_value':np.nanmean})
```

```python
df.groupby('cancellation_policy').agg({'review_scores_value':(np.nanmean,np.nanstd),
                                       'reviews_per_month':np.nanmean})
```

### Transformation
Transformation is different from aggregation. Where agg() returns a single value per column, so one row per
group, tranform() returns an object that is the same size as the group. Essentially, it broadcasts the
function you supply over the grouped dataframe, returning a new dataframe. This makes combining data later
easy.

```python
cols = ['cancellation_policy', 'review_scores_value']
df[cols].head()
```

```python
transform_df = df[cols].groupby('cancellation_policy').transform(np.nanmean)
transform_df.head()
```

```python
transform_df.rename({'review_scores_value':'mean_review_scores'},axis='columns',
                    inplace=True) # renamed the column
transform_df.head()
```

```python
df = df.merge(transform_df, left_index=True, right_index=True)
df.head()
```

```python
df['mean_diff']=np.absolute(df['review_scores_value']-df['mean_review_scores'])
df.head()
```

### Filtering

```python
df.groupby('cancellation_policy').filter(lambda x: np.nanmean(x['review_scores_value'])>9.2)
```

### Applying

```python
df = pd.read_csv('listings.csv')
df = df[['cancellation_policy', 'review_scores_value']]
df.head()
```

```python
def calc_mean_review_scores(group):
    avg=np.nanmean(group['review_scores_value'])
    group['review_scores_mean'] = np.abs(avg-group['review_scores_value'])
    return group

df.groupby('cancellation_policy').apply(calc_mean_review_scores).head()
```
