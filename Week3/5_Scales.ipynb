```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D'],
                  index=['excellent']*3 + ['good']*3 + ['ok']*3 + ['poor']*2,
                  columns=['Grades'])
df
```

```python
df.dtypes
```

```python
df['Grades'].astype('category')
```

```python
l=['A+', 'A', 'A-', 'B+', 'B', 'B-', 'C+', 'C', 'C-', 'D+', 'D']
l.reverse()
l
```

```python
my_categories = pd.CategoricalDtype(categories=['D', 'D+', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A', 'A+'],
                                    ordered=True)
grades=df['Grades'].astype(my_categories)
grades
```

```python
df[df['Grades']>'C'] # lexicographical comparision
# note - df is not of categorical type
```

```python
grades[grades>'C']
# note - grades is of categorical type
```
