Python programmers will often suggest that there many ways the language can be used to solve a particular problem. But that some are more appropriate than others. The best solutions are celebrated as Idiomatic Python.  
A sort of sub-language within Python, Pandas has its own set of idioms. We've alluded to some of these already, such as using vectorization whenever possible, and not using iterative loops if you don't need to. Several developers and users within the Panda's community have used the term __pandorable__ for these idioms.

```python
import pandas as pd
import numpy as np
import timeit
```

```python
df = pd.read_csv('census.csv')
df.head()
```

```python
#pandorable
(df.where(df['SUMLEV']==50)
   .dropna()
   .set_index(['STNAME','CTYNAME'])
   .rename(columns={'ESTIMATESBASE2010':'Estimates Base 2010'}))
```

```python
#non-pandorable
df = df[df['SUMLEV']==50]
df.set_index(['STNAME','CTYNAME'], inplace=True)
df.rename(columns={'ESTIMATESBASE2010':'Estimates Base 2010'})
```

TIMEIT pandorable vs non-pandorable :-

```python
#pandoarable
def first_approach():
    global df
    return (df.where(df['SUMLEV']==50)
            .dropna()
            .set_index(['STNAME','CTYNAME'])
            .rename(columns={'ESTIMATESBASE2010':'Estimates Base 2010'}))

df = pd.read_csv('census.csv')
timeit.timeit(first_approach, number=10) # number=10 no. of times
```

```python
#non-pandorable
def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME','CTYNAME'],inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010':'Estimates Base 2010'})

df = pd.read_csv('census.csv')
timeit.timeit(second_approach, number=10) # number=10 no. of times
```

**Conclusion** - What appears to be stylistic idioms might have performance issues that you need to consider as well.

Lets find min and max along columns of population estimate from 2011 to 2015 -

```python
df[['POPESTIMATE2011',
    'POPESTIMATE2012',
    'POPESTIMATE2013',
    'POPESTIMATE2014',
    'POPESTIMATE2015']]
```

```python
def min_max(row):
    data = row[['POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min':np.min(data),'max':np.max(data)})

df.apply(min_max,axis=1)
```

Another way adding min max columns to dataframe itself -

```python
def min_max(row):
    data = row[['POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row

df.apply(min_max, axis=1).head()[['POPESTIMATE2011',
                                  'POPESTIMATE2012',
                                  'POPESTIMATE2013',
                                  'POPESTIMATE2014',
                                  'POPESTIMATE2015', 'max', 'min']]
```

Using Lambda - 

```python
rows = ['POPESTIMATE2011', 'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014', 'POPESTIMATE2015']
df.apply(lambda x:pd.Series({'max':np.max(x[rows]),'min':np.min(x[rows])}), axis=1)
```
